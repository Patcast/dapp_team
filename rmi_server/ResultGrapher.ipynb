# Import libraries

```python
import numpy as np
from numpy import genfromtxt

import matplotlib.pyplot as plt
```

# Get data from CSV

```python
my_data = genfromtxt('result.csv', delimiter = ',')
```

# Calculate mean and standard deviation

```python
my_data = my_data[1:, :]    # remove the first row (heading)
my_data[:, 1] /= 1000
# group the data by the first dimension
unique_x = np.unique(my_data[:, 0])

# calculate the mean and standard deviation for each group
means = []
stds = []
for x in unique_x:
    idx = np.where(my_data[:, 0] == x)
    group = my_data[idx, 1]
    means.append(np.mean(group))
    stds.append(np.std(group))
```

# Draw graph

```python
plt.errorbar(unique_x, means, yerr=stds, fmt='-o')
plt.xlabel('number of rooms')
plt.xscale('log', base=2)
plt.ylabel('latency (ms)')
plt.show()
```
